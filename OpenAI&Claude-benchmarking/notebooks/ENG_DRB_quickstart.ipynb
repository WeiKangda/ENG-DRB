```python
from eng_drb_benchmark.data import load_eng_drb, summarize_relation_counts, export_gold_jsonl

dataset_dict = load_eng_drb()
split_name = next(iter(dataset_dict.keys()))
dataset = dataset_dict[split_name]
print('Split:', split_name)
print('Summary:', summarize_relation_counts(dataset))
```
